```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('worldcup_alltime_table.csv')

df.plot(x='Nation',y=['Titles','Runners-up'],kind='bar',color=['red','blue'])

plt.xlabel('Nations')
plt.ylabel('World-Cups and runners-up')
plt.title('Bar Graph showing World-Cup winners and runners-up')

plt.show()

```

```python
#pie-chart to show the most points accumulated by top teams

df.plot(kind='pie',y='Pts',labels=df['Nation'],autopct='%1.2f%%',legend=False)
plt.title("Pie chart showing the points accumulated by the top 10 world-cup teams")
plt.show()
```

```python
df.plot(kind='bar',y=['GF','GA'],x='Nation',color=['Teal','Cyan'])
plt.ylabel('Goals Scored and Conceded')
plt.xlabel('Nations')

plt.show()
```
